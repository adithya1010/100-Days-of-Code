```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

```

```python
# Load the iris dataset
iris = datasets.load_iris()
X = iris.data[:, 2:]  # We'll only use petal length and width for simplicity
y = iris.target
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create and train the SVM model
svm_model = SVC(kernel='linear', C=1) # Using a linear kernel for simplicity.  C is the regularization parameter.
svm_model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = svm_model.predict(X_test)
```

```python
# Evaluate the model
print(classification_report(y_test, y_pred, target_names=iris.target_names))
print(confusion_matrix(y_test, y_pred))
```

```python
#Example prediction (optional)
example_petal = np.array([[5.1, 2.4]]) #Example Petal Length and width
predicted_species = svm_model.predict(example_petal)
print(f"\nExample prediction: Petals at {example_petal} predict species {iris.target_names[predicted_species[0]]}")
```
